# how to get live market data in minutes

```python
#Interval required 5 minutes
data = yf.download(tickers='AAPL', period='1d', interval='1m')
data2 = yf.download(tickers='NFLX', period='1d', interval='1m')

#Print data
data
data2
```

```python
#Interval required 5 minutes
data = yf.download(tickers='AAPL', period='1d', interval='1m')
data2 = yf.download(tickers='NFLX', period='1d', interval='1m')
#declare figure
fig=go.Figure()

fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

fig.update_layout(
    title=' Live Share Price AAPL:',
    yaxis_title='Stock Price (USD per Shares)')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
            ])
    )
)
fig.show()

fig=go.Figure()

fig.add_trace(go.Candlestick(x=data.index,
                open=data2['Open'],
                high=data2['High'],
                low=data2['Low'],
                close=data2['Close'], name = 'market data'))

fig.update_layout(
    title=' Live Share Price NFLX:',
    yaxis_title='Stock Price (USD per Shares)')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
           dict(step="all")
            ])
        )
)

fig.show()


```
